```python
url = "https://api.harvardartmuseums.org/object"
params = {
            "apikey": API_KEY,
            "size": 100,
            "classification": "Vessels",
            "page": 1
        }

total_records_to_collect = 2500
collected_records = []
page = 1

while len(collected_records) < total_records_to_collect:
    params["page"] = page
    response = requests.get(url, params=params)
    data = response.json()
    records = data.get('records', [])

    if not records:
        break  # No more records to fetch

    collected_records.extend(records)
    page += 1

data = {'records': collected_records} # Update data with all collected records
```

```python
# Fetch the data from the API again to ensure it's available
import requests

API_KEY = "432b9513-8af3-4b52-b27b-f1d4235ea583"

url = "https://api.harvardartmuseums.org/object"
params = {
            "apikey": API_KEY,
            "size": 100,
            "page": 1,  # scroll through pages and collect 2500 records
            "classification": "Vessels"
        }
response = requests.get(url, params=params)
data = response.json()


# Function to insert data into artifact_metadata table
def insert_metadata(mycursor, data):
    for record in data['records']:
        mycursor.execute("""
            INSERT INTO artifact_metadata (id, title, culture, period, century, medium, dimensions, description, department, classification, accessionyear, accessionmethod)
            VALUES (%s,%s,%s,%s,%s,%s,%s,%s, %s, %s, %s, %s)

        """, (
            record.get('id'),
            record.get('title'),
            record.get('culture'),
            record.get('period'),
            record.get('century'),
            record.get('medium'),
            record.get('dimensions'),
            record.get('description'),
            record.get('department'),
            record.get('classification'),
            record.get('accessionyear'),
            record.get('accessionmethod')
        ))

# Function to insert data into artifact_media table
def insert_media(mycursor, data):
    for record in data['records']:
        mycursor.execute("""
            INSERT INTO artifact_media (objectid, imagecount, mediacount, colorcount, rank, datebegin, dateend)
            VALUES (%s, %s, %s, %s, %s, %s, %s)
        """, (
            record.get('objectid'),
            record.get('imagecount'),
            record.get('mediacount'),
            record.get('colorcount'),
            record.get('rank'),
            record.get('datebegin'),
            record.get('dateend')
        ))

# Function to insert data into artifact_colors table
def insert_colors(mycursor, data):
    for record in data['records']:
        colors_list = record.get('colors')
        if colors_list:
            for color_detail in colors_list:
                mycursor.execute("""
                    INSERT INTO artifact_colors (objectid, color, spectrum, hue, percent, css3)
                    VALUES (%s, %s, %s, %s, %s, %s)
                """, (
                    record.get('objectid'),
                    color_detail.get('color'),
                    color_detail.get('spectrum'),
                    color_detail.get('hue'),
                    color_detail.get('percent'),
                    color_detail.get('css3')
                ))

# mydb.commit()
# print("Data inserted successfully.")
```
